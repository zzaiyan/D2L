```python
import pandas as pd
import numpy as np
import ast
from tqdm import tqdm
```

```python
df_job = pd.read_excel('./position.xlsx')
df_job.shape
```

```python
df_per = pd.read_excel('./user.xlsx')
df_per.shape
```

```python
df_job.head()
```

```python
行业统计 = {}
```

```python
for i in range(len(df_job)):
  hy_txt = df_job.loc[i,'行业']
  hy_list = ast.literal_eval(hy_txt)
  for hy in hy_list:
    if hy in 行业统计:
      行业统计[hy] += 1
    else:
      行业统计[hy] = 1
```

```python
s = pd.Series(行业统计).sort_values(ascending=False)
s.to_excel('./行业统计.xlsx')
```
