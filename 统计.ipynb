```python
行业统计 = {}
```

```python
for i in range(len(df_job)):
  hy_txt = df_job.loc[i,'行业']
  hy_list = ast.literal_eval(hy_txt)
  for hy in hy_list:
    if hy in 行业统计:
      行业统计[hy] += 1
    else:
      行业统计[hy] = 1
```

```python
s = pd.Series(行业统计).sort_values(ascending=False)
s.to_excel('./行业统计.xlsx')
```

```python
学历统计 = {'None':0}
```

```python
for i in range(len(df_per)):
  hy_txt = df_per.loc[i,'教育经历']
  hy_list = ast.literal_eval(hy_txt)
  if hy_list == []:
    学历统计['None'] += 1
    continue
  学历 = hy_list[0]['educationalBackground']
  if 学历 in 学历统计:
    学历统计[学历] += 1
  else:
    学历统计[学历] = 1
```

```python
期望工资统计 = {}
```

```python
for i in range(len(df_per)):
  high = df_per.loc[i,'预期最大薪资']
  low = df_per.loc[i,'预期最小薪资']
  avl = (high+low)/2
  x = int(avl/1000)*1000
  print(x)
  if x in 期望工资统计:
    期望工资统计[x] += 1
  else:
    期望工资统计[x] = 1
```
